# Stationary testbed (exercise 2.3)

```python
# Create bandit problems with k=10, and each action created using
# a mean expected reward of 0, with a variance of 1. A variance of 1
# is used when selecting from that action, based off its expected reward

problems = [
    BanditProblem(10, 0, 1, 1)
    for _ in range(2000)
]
```

```python
# Average maximum reward

sum([problem.action_reward_means[problem.optimal_action]for problem in problems]) / 2000
```

```python
# Train agents on the bandit problems, varying the explore fraction 

data = {
    "0": run_testbed(problems, 10, 0.0, 0.0, 1000),
    "0.01": run_testbed(problems, 10, 0.01, 0.0, 1000),
    "0.1": run_testbed(problems, 10, 0.1, 0.0, 1000),
}
```

```python
# Process data from trained agents to obtain mean rewards
# across the agent population

e0 = []
e1 = []
e2 = []

for t in range(1000):
    e0.append(
        sum(
            a.rewards_received[t] for a in data["0"]
        ) / 2000
    )
    e1.append(
        sum(
            a.rewards_received[t] for a in data["0.1"]
        ) / 2000
    )
    e2.append(
        sum(
            a.rewards_received[t] for a in data["0.01"]
        ) / 2000
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1000)

plt.figure(figsize=(15,10))

plt.plot(x, e0)
plt.plot(x, e1)
plt.plot(x, e2)

plt.legend(['e = 0', 'e = 0.1', 'e = 0.01'], loc='lower right')

plt.show()
```

```python
# Find % optimal action

o0 = []
o1 = []
o2 = []

for t in range(1000):
    o0.append(
        sum(
            a.optimal_action[t] for a in data["0"]
        ) / 2000
    )
    o1.append(
        sum(
            a.optimal_action[t] for a in data["0.1"]
        ) / 2000
    )
    o2.append(
        sum(
            a.optimal_action[t] for a in data["0.01"]
        ) / 2000
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1000)

plt.figure(figsize=(15,10))

plt.plot(x, o0)
plt.plot(x, o1)
plt.plot(x, o2)

plt.legend(['e = 0', 'e = 0.1', 'e = 0.01'], loc='lower right')

plt.show()
```

# Non-stationary testbed (exercise 2.5)

```python
# Non-stationary version
from ten_armed_testbed import run_testbed, run_testbed_constant_alpha, BanditProblem

non_stationary_problems = [
    BanditProblem(10, 0, 0, 1, stationary=False)
    for _ in range(2000)
]
```

```python
# Train agents on the non-stationary problems

non_stationary_data = {
    "sample-average": run_testbed(non_stationary_problems, 10, 0.1, 0.0, 10000),
    "constant": run_testbed_constant_alpha(non_stationary_problems, 10, 0.1, 0.0, 10000)
}
```

```python
# Process data from trained agents to obtain mean rewards
# across the agent population

ns0 = []
ns1 = []

for t in range(10000):
    ns0.append(
        sum(
            a.rewards_received[t] for a in non_stationary_data["sample-average"]
        ) / 2000
    )
    ns1.append(
        sum(
            a.rewards_received[t] for a in non_stationary_data["constant"]
        ) / 2000
    )
    
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10000)

plt.figure(figsize=(15,10))

plt.plot(x, ns0)
plt.plot(x, ns1)

plt.legend(['sample average', 'constant step size'], loc='lower right')

plt.show()
```

```python
sum([p.action_reward_means[p.optimal_action] for p in non_stationary_problems]) / len(non_stationary_problems)
```

```python
# Find % optimal action

nso0 = []
nso1 = []

for t in range(10000):
    nso0.append(
        sum(
            a.optimal_action[t] for a in non_stationary_data["sample-average"]
        ) / 2000
    )
    nso1.append(
        sum(
            a.optimal_action[t] for a in non_stationary_data["constant"]
        ) / 2000
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10000)

plt.figure(figsize=(15,10))

plt.plot(x, nso0)
plt.plot(x, nso1)



plt.legend(['sample average', 'constant step size'], loc='lower right')

plt.show()
```
